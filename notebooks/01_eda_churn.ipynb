```python
import pandas as pd
import os

# Set up correct absolute path
base_path = 'C:/Users/Admin/Desktop/churn/data/'
save_path = os.path.join(base_path, 'data_cleaned/')
os.makedirs(save_path, exist_ok=True)

# Utility function to clean and save CSVs
def clean_and_save(df, name):
    try:
        df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
        df.to_csv(os.path.join(save_path, f"{name}_clean.csv"), index=False)
        print(f"✅ {name} cleaned and saved.")
    except Exception as e:
        print(f"❌ Error with {name}: {e}")

# 1. Amazon Sale Report
try:
    amazon = pd.read_csv(base_path + 'Amazon Sale Report.csv', low_memory=False)
    amazon.columns = amazon.columns.str.lower().str.replace(' ', '_')
    amazon['date'] = pd.to_datetime(amazon['date'], errors='coerce')
    amazon.drop(columns=['asin', 'courier_status'], inplace=True, errors='ignore')
    amazon.to_csv(save_path + 'amazon_clean.csv', index=False)
    print("✅ Amazon Sale Report cleaned and saved.")
except Exception as e:
    print("❌ Amazon Sale Report error:", e)

# 2. International Sale Report
try:
    intl = pd.read_csv(base_path + 'International sale Report.csv')
    intl.columns = intl.columns.str.strip().str.lower().str.replace(" ", "_")
    intl['date'] = pd.to_datetime(intl['date'], errors='coerce')
    for col in ['pcs', 'rate', 'gross_amt']:
        if col in intl.columns:
            intl[col] = pd.to_numeric(intl[col], errors='coerce')
    clean_and_save(intl, 'intl')
except Exception as e:
    print("❌ International Sale Report error:", e)

# 3. Cloud Warehouse Comparison Chart
try:
    warehouse = pd.read_csv(base_path + 'Cloud Warehouse Compersion Chart.csv')
    clean_and_save(warehouse, 'warehouse')
except Exception as e:
    print("❌ Warehouse error:", e)

# 4. Sale Report
try:
    sale = pd.read_csv(base_path + 'Sale Report.csv')
    clean_and_save(sale, 'sale')
except Exception as e:
    print("❌ Sale Report error:", e)

# 5. P & L March 2021
try:
    pl = pd.read_csv(base_path + 'P  L March 2021.csv')  # Make sure filename is correct
    clean_and_save(pl, 'pl_march')
except Exception as e:
    print("❌ P & L March 2021 error:", e)

# 6. May-2022
try:
    may = pd.read_csv(base_path + 'May-2022.csv')
    clean_and_save(may, 'may_2022')
except Exception as e:
    print("❌ May-2022 error:", e)

# 7. Expense IIGF
try:
    expense = pd.read_csv(base_path + 'Expense IIGF.csv')
    clean_and_save(expense, 'expense')
except Exception as e:
    print("❌ Expense IIGF error:", e)

print("\n✅✅ All processing complete. Please check 'data_cleaned' folder.")
```

```python
import pandas as pd
import datetime as dt

# Load and prepare Amazon cleaned data
amazon = pd.read_csv('C:/Users/Admin/Desktop/churn/data_cleaned/Amazon Sale Report_clean.csv')

# Convert data types
amazon['date'] = pd.to_datetime(amazon['date'], errors='coerce')
amazon['amount'] = pd.to_numeric(amazon['amount'], errors='coerce')
amazon_cleaned = amazon.dropna(subset=['date', 'amount'])

# Reference date for Recency calculation
ref_date = amazon_cleaned['date'].max() + pd.Timedelta(days=1)

# Group by SKU to compute RFM
rfm_df = amazon_cleaned.groupby('sku').agg({
    'date': lambda x: (ref_date - x.max()).days,   # Recency
    'sku': 'count',                                # Frequency
    'amount': 'sum'                                # Monetary
}).rename(columns={
    'date': 'recency',
    'sku': 'frequency',
    'amount': 'monetary'
}).reset_index()

# Show top 10 RFM records
print(rfm_df.sort_values(by='monetary', ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Assuming `rfm_df` is your RFM dataframe from earlier
# Histogram for each RFM metric
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(rfm_df['recency'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Recency Distribution')

sns.histplot(rfm_df['frequency'], bins=20, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Frequency Distribution')

sns.histplot(rfm_df['monetary'], bins=20, kde=True, ax=axes[2], color='salmon')
axes[2].set_title('Monetary Value Distribution')

plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(rfm_df[['recency', 'frequency', 'monetary']].corr(), annot=True, cmap='coolwarm')
plt.title('RFM Correlation Heatmap')
plt.show()

# Optional: Scatter plots for quick insight
sns.pairplot(rfm_df[['recency', 'frequency', 'monetary']])
plt.suptitle('RFM Scatter Matrix', y=1.02)
plt.show()
```

```python
# Step 1: Assign Quantile Scores
rfm_df['R_Score'] = pd.qcut(rfm_df['recency'], 4, labels=[4, 3, 2, 1])
rfm_df['F_Score'] = pd.qcut(rfm_df['frequency'], 4, labels=[1, 2, 3, 4])
rfm_df['M_Score'] = pd.qcut(rfm_df['monetary'], 4, labels=[1, 2, 3, 4])

# Step 2: Combine RFM Scores
rfm_df['RFM_Segment'] = rfm_df['R_Score'].astype(str) + rfm_df['F_Score'].astype(str) + rfm_df['M_Score'].astype(str)
rfm_df['RFM_Score'] = rfm_df[['R_Score', 'F_Score', 'M_Score']].sum(axis=1).astype(int)

# Step 3: Define Segment Labels
def segment_customer(row):
    if row['RFM_Score'] >= 9:
        return 'Champions'
    elif row['R_Score'] == 4:
        return 'Loyal'
    elif row['F_Score'] == 4:
        return 'Frequent Buyer'
    elif row['M_Score'] == 4:
        return 'Big Spender'
    elif row['R_Score'] == 1:
        return 'At Risk'
    else:
        return 'Others'

rfm_df['Segment'] = rfm_df.apply(segment_customer, axis=1)

# Preview
rfm_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# 1. Countplot of Segment Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=rfm_df, x='Segment', order=rfm_df['Segment'].value_counts().index, palette='Set2')
plt.title('Customer Segments Distribution')
plt.xticks(rotation=45)
plt.ylabel('Number of SKUs')
plt.tight_layout()
plt.show()

# 2. Heatmap of Mean RFM Scores per Segment
segment_scores = rfm_df.groupby('Segment')[['recency', 'frequency', 'monetary']].mean().round(1)

plt.figure(figsize=(10, 6))
sns.heatmap(segment_scores, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Average R, F, M Values by Segment')
plt.tight_layout()
plt.show()

# 3. Boxplot of Monetary by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(data=rfm_df, x='Segment', y='monetary', palette='coolwarm', order=rfm_df['Segment'].value_counts().index)
plt.title('Monetary Distribution by Segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Boxplot of Frequency by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(data=rfm_df, x='Segment', y='frequency', palette='viridis', order=rfm_df['Segment'].value_counts().index)
plt.title('Frequency Distribution by Segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Tag churn risk based on segment
def churn_status(segment):
    if segment in ['At Risk', 'Hibernating', 'Lost']:
        return 'Churn Likely'
    elif segment in ['Loyal', 'Champion', 'Potential Loyalist']:
        return 'Retained'
    else:
        return 'Needs Nurturing'

rfm_df['Churn_Status'] = rfm_df['Segment'].apply(churn_status)

# View value counts
print(rfm_df['Churn_Status'].value_counts())
```
